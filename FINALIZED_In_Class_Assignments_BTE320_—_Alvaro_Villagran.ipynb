To calculate taxes, multiply gross profit by tax rate, and to get net income you subtract taxes from gross profit

Recursive function: string reversal
- create a recursive function that takes as input a string and returns it reversed

- exaples:
  - revese('hello') -> 'olleh'
  - reverse('12345') -> '54321'

```python
def reverse(s):
  if len(s) > 1:
    #return s[-1] + reverse(s[:-1])
    #return reverse(s[1:]) + s[0]
    return s[-1] + reverse(s[1:-1]) + s[0]
  else:
    return s

reverse('hello')
```

A quality control manager would like to find the average weight of a list of weights. Prior experience indicates that the first two measure are atypical, and the practie is to exlude these from the average calculation.

Write a program that calulates the average of a list of numbers that given as inputs from the user, excluding the first two values.

For your program you must define and use the following functions:
- getValues(n): Prompt the use for n values, returns the list.
- calcAverage(alist): Calculates the average from the list.

Difference with Assignment II: use try-except to avoid invalid inputs(negative values, strings), invalid operations, etc.

```python
def getValues(n: int) -> list:
  '''
  Creates a list of weights. The eights are given by the user
  Input:
  - number of items in the list
  Output:
  - a list of weights
  '''
  items = []

  for i in range(n):
    try:
      item = float(input('Enter weight value: '))
    except Exception as e:
      print(e)
      item = float('nan')
    items.append(item)

  return items

def calcAverage(alist: list) -> float:
  '''
  Input:
  - A list of weights
  Output:
  - Average value; excludes the first two items
  '''
  try:
    return sum(alist[2:]) / len(alist[2:])
  except ZeroDivisionError:
    print('The list is empty!')
    return float('nan')

def main():
  num = int(input('Enter number of weights: '))

  if num <= 0:
    raise ValueError('num must be positive')

  weights = getValues(num)
  avg = calcAverage(weights)
  print(f'Average weight: {avg} lbs.')

main()
```

LOAN & npv MINI TOOL KIT

```python
def loan_payment(principal: float, annual_rate: float, months: int) -> float:
    assert isinstance(principal, (int, float)), "principal must be numeric"
    assert isinstance(annual_rate, (int, float)), "annual_rate must be numeric"
    assert isinstance(months, int), "months must be an integer"

    if principal <= 0:
        raise ValueError("Principal must be > 0.")
    if annual_rate < 0:
        raise ValueError("Annual rate cannot be negative.")
    if months <= 0:
        raise ValueError("Number of months must be > 0.")

    r = annual_rate / 12.0
    n = months
    P = float(principal)

    if r == 0.0:
        return P / n
    else:
        return r * P / (1.0 - (1.0 + r) ** (-n))


if __name__ == "__main__":
    try:
        P = float(input("Enter principal: "))
        apr = float(input("Enter annual interest rate as a decimal: "))
        n = int(input("Enter number of months: "))
        payment = loan_payment(P, apr, n)
        print(f"Monthly payment: {payment:.2f}")
    except Exception as e:
        print(e)
```

Part A - compute simple ROI

```python
def roi(cost: float, revenue: float) -> float:
  assert cost > 0, "The cost must be positive!"
  assert isinstance(cost, (int, float)), "cost must be numerical"
  assert isinstance(revenue, (int, float)), "revenue must be numerical"

  return (revenue - cost) / cost

if __name__ == "__main__":
  c = float(input('Enter cost value: '))
  r = float(input('Enter revenue value: '))
  retInvest = roi(c, r)
  print(retInvest)
```

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

Create a class Point that has:
- two coordinate attributes (x, y)
- A method distance() that calculates the distance between two instances of the class

Create a class rectangle that has:
 - six attributes: four Point instances and two attributes for width/height. The latter two are calvulated based on pairs of the four Point instances.
 - Three mthods for calculting width, height, area of the rectangle.

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y


  def distance(self, other):
    x_diff_sq = (self.x - other.x)**2
    y_diff_sq = (self.y - other.y)**2
    return (x_diff_sq + y_diff_sq)**0.5

class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(1,0)
    self.p3 = Point(1,1)
    self.p4 = Point(0,1)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p4)

  def area(self):
   return self.width * self.height


sq = Rectangle()
sq.area()
```
